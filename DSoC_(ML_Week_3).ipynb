# **K-Means** **Clustering**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
X,y= make_blobs(n_samples=1000,centers=3,n_features=2,random_state=101)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
# Manual method to find K-Value
# Elbow Method
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
import seaborn as sns
sns.lineplot(x=range(1,11),y=wcss)
# We find K=3
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_labels=kmeans.fit_predict(X_train)
y_labels
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels)
```

```python
y_test_labels=kmeans.predict(X_test)
y_test_labels
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test_labels)
```

```python
# Automated K finding method
! pip install kneed
```

```python
from kneed import KneeLocator
kl=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
kl.elbow
```

```python
# Silhouette Score
from sklearn.metrics import silhouette_score
silhouette_coeff=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    score=silhouette_score(X_train,kmeans.labels_)
    silhouette_coeff.append(score)
silhouette_coeff
```

```python
plt.plot(range(2,11),silhouette_coeff)
plt.xticks(range(2,11))
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
```

# **Hierarchial Clustering**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
iris=datasets.load_iris() #Import IRIS dataset
```

```python
iris_data=pd.DataFrame(iris.data)
iris_data
```

```python
iris_data.columns=iris.feature_names
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(iris_data)
#X_scaled=pd.DataFrame(X_scaled)
X_scaled.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(X_scaled)
pca_scaled.shape
pca_scaled
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
# Agglomerative Clustering
# To Construct a Dendogram
import scipy.cluster.hierarchy as sc
# Plot a dendogram
plt.figure(figsize=(20,7))
plt.title("Dendrograms")

# Create Dendogram
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))

plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
```

```python
## We find k=2 from above figure
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_coeff=[]

for k in range(2,11):
    agglo=AgglomerativeClustering(n_clusters=k,affinity='euclidean',linkage='ward')
    agglo.fit(X_scaled)
    score=silhouette_score(X_scaled,agglo.labels_) # Should I use X_scaled or pca_scaled here
    sil_coeff.append(score)
```

```python
plt.plot(range(2,11),sil_coeff)
plt.xticks(range(2,11))
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
```

# **DBSCAN Clustering**

```python
X,y=make_moons(n_samples=100,noise=0.07)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_sca=scaler.fit_transform(X)
```

```python
plt.scatter(X_sca[:,0],X_sca[:,1],c=dbscan.labels_)
```
